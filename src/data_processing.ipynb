```python
import pandas as pd
import numpy as np
```

```python
df_1room = pd.read_csv('../data/raw/1_2025-05-11_12-38.csv')
df_2room = pd.read_csv('../data/raw/2_2025-05-11_13-27.csv')
df_3room = pd.read_csv('../data/raw/3_2025-05-11_13-32.csv')
df_4room = pd.read_csv('../data/raw/4_2025-05-11_13-35.csv')

df = pd.concat([df_1room, df_2room, df_3room, df_4room], ignore_index=True)
```

```python
df
```

```python
df['rooms_count'].value_counts()
```

```python
df = df[df['rooms_count'] != -1]
```

```python
#т.к. данных немного, и мы рассматриваем только однокомнатные квартиры в Москве на продажу, можно удалить следующие столбцы
columns_to_keep = ['rooms_count', 'floor', 'floors_count', 'total_meters', 'price']
df = df[columns_to_keep]
```

```python
df.to_csv('../data/processed/data_11_05.csv')
```
